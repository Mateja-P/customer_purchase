```python
import pandas as pd
from matplotlib import pyplot as plt 
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score

#Classification problem

#No missing values
#Wrong data types
    #Gender
    #Review
    #Education
    #Purchase
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df = pd.get_dummies(df, columns=['gender', 'review', 'education', 'purchased'], drop_first=True)
df.info()
```

```python
plt.figure(figsize=(12,9))
sns.heatmap(
    data=df.corr(),     
    cmap=plt.cm.RdBu,
    vmax=1.0,
    linewidths=0.1,
    linecolor='white',
    square=True,
    annot=True
)
```

```python
sns.jointplot(x='education_UG', y='purchased_Yes', data=df)
```

```python
X = df.loc[:, df.columns!= 'purchased_Yes']
y = df['purchased_Yes']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
#LogisticRegression

lr = LogisticRegression()

lr.fit(X_train, y_train)
lr.fit(X_test, y_test)

y_lr_pred_train = lr.predict(X_train)
y_lr_pred_test = lr.predict(X_test)
```

```python
accuracy_score(y_train, y_lr_pred_train)
```

```python
accuracy_score(y_test, y_lr_pred_test)
```

```python
#LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()

lda.fit(X_train, y_train)
lda.fit(X_test, y_test)

y_lda_pred_train = lda.predict(X_train)
y_lda_pred_test = lda.predict(X_test)
```

```python
accuracy_score(y_train, y_lda_pred_train)
```

```python
accuracy_score(y_test, y_lda_pred_test)
```

```python
#KNN 

knn = KNeighborsClassifier(n_neighbors=3)

scaler_train = StandardScaler().fit(X_train)
scaler_test = StandardScaler().fit(X_test)

X_s_train = scaler_train.fit_transform(X_train)
X_s_test = scaler_test.fit_transform(X_test)

knn.fit(X_s_train, y_train)
knn.fit(X_s_test, y_test)

y_knn_pred_train = knn.predict(X_s_train)
y_knn_pred_test = knn.predict(X_s_test)
```

```python
accuracy_score(y_train, y_knn_pred_train)
```

```python
accuracy_score(y_test, y_knn_pred_test)
```
